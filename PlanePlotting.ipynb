```python
# Create the figure
plt.clf()
fig = plt.figure()

# Add an axes
ax = fig.add_subplot(111,projection='3d')

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a,b,c,d = 1,2,3,4

x = np.linspace(-100,100,10)
y = np.linspace(-100,100,10)

X,Y = np.meshgrid(x,y)
Z = (d - a*X - b*Y) / c

ax = fig.gca(projection='3d')
ax.axes.set_xlim3d(left=-100, right=100)
ax.axes.set_ylim3d(bottom=-100, top=100)
ax.axes.set_zlim3d(bottom=-100, top=100)

# and plot the point 
point2= np.array([[ 3.62317578, 90.39276123, 51.40513229],
                  [-68.00674184, -90.61721802, -65.40477498],
                  [-28.52493487,-3.69403103,-5.05969256]])


ax.scatter(point2[0] , point2[1] , point2[2],  color='green')
surf = ax.plot_surface(X, Y, Z)
```
